```python
#Tutorial 1 - Goal Keeper problem

import pandas as pd 
import sasoptpy as so
```

```python
df = pd.read_csv('../data/input.csv')
df.head()
```

```python
gk_data = df[df['element_type'] == 1].copy().reset_index() #create a set of data with only the GK position. It will the reset the index using reset()
gk_data.head() 
```

```python
# Now we define the problem 

model = so.Model(name = 'gk_model') #this is the obkject that will include evert detail from out problem

#now we create an index set 
gk_data.set_index('index', inplace=True)


gk_data.head()
```

```python
players = gk_data.index.tolist() #this creates an indeci list of the player ID for each GK
players
```

```python
lineup = model.add_variables(players, name='lineup', vartype =so.binary) #creates a varibale dictionary of each players name and a binary number (1 = selected, 0 = not selected)



bench = model.add_variables(players, name='bench', vartype = so.binary) #cretaes the same thing for a bench GK

```

```python
#get expected points for next week --> multiplies either 1 or 0 with expected points (THIS IS OUR OBJECTVIE)
total_xp = so.expr_sum(lineup[p] * gk_data.loc[p, 'ep_next'] for p in players) + 0.1 * so.expr_sum(bench[p] * gk_data.loc[p, 'ep_next'] for p in players)    # creates for the total expected points of your selected lineup, which the optimizer can maximize.
```

```python
total_xp #we can now see each players id with their respective expected points --> the bench is 10% of each player 
```

```python
model.set_objective(total_xp, name = 'total_xo_obj', sense=so.maximize) #sets the model to maximise mode 
```

```python
#so we need to write some constraints due to the rules of FPL (this is /(s) as it will parse over multiple players)

model.add_constraints((lineup[p] + bench[p] <= 1 for p in players), name ='lineup_or_bench');
```

```python
# we add a single containt becuase we can only have a single lineup goalkeeper 
model.add_constraint(so.expr_sum(lineup[p] for p in players) ==1, name ='single_lineup');
```

```python
#we need to add a contraint so that we only select 1 keeper on the bench 
model.add_constraint(so.expr_sum(bench[p] for p in players) == 1, name='single_bench')
```

```python
#add constraint for our budget

budget = 9.2 
#df['BV'] = df['now_cost'] /10 
model.add_constraint(so.expr_sum((lineup[p] + bench[p]) * (gk_data.loc[p, 'now_cost'] / 10) for p in players) <= budget, name='budget_con')
```

```python
print(model.export_mps(fetch=True)) #this exports the optimization moduel to a MPS (Mathematical programming syntax) language used in linear programming 
```

```python
#now we are ready for the proble. We need to be able to give this formulation to a slover CBC in our case. 
#we need to print this above object into the system solver 
```

```python
model.set_objective(-total_xp, name='total_xp_obj',sense='N')
```

```python
model.export_mps(filename="gk.mps") #creates a way to represent the problem 
```

```python
with open ('solution.txt', 'r') as f:
    for v in model.get_variables():
        v.set_value(0)
    for line in f:
        if 'objective value' in line:
            continue 
        words = line.split()
        var = model.get_variable(words[1])
        var.set_value(float(words[2]))
```
